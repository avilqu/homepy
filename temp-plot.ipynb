```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
plt.style.use('classic')
```

```python
import csv
```

```python
import config as cfg
```

```python
x = []
y1 = []
y2 = []
```

```python
with open(f'{cfg.DATA_DIR}2021-08-24.txt') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        data_count = 0
        for row in csv_reader:
            x.append(row[0])
            y1.append(row[1])
            y2.append(row[2])
            data_count += 1
```

```python
plt.plot(x, y1, y2)
plt.show()
```
